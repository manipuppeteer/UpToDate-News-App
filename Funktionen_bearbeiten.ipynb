```python
import requests

api_key = "ecb9e858a7fb47a6aa7b33d668c1ba6e"
base_url = "https://newsapi.org/v2/top-headlines"

def get_news (category: str )-> list[dict]:

    params = {
        'country': 'us',
        'category': category.lower(),
        'apiKey': api_key
    }

    reponse = requests.get(base_url,params=params)
    if reponse.status_code == 200:
        data = reponse.json()
    return data.get('articles', 1)
```

```python
get_news('sports')
```

```python
def such_articles(category:str, word:str) ->[list[dict]]:

    params = {
        'category': category.lower(),
        'q': word,
        'apiKey': api_key
    }

    response = requests.get(base_url,params=params)
    if response.status_code == 200:
        data = response.json()
    return data.get('articles', 1)
```
